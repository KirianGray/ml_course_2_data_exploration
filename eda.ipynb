# Data quality assessment / Оценка качества данных

```python
#Выведем и посмотрим полную сводку о составе датафрейма и используемых типах данных
titanic_data.info()
#Видим, что количество элементов в столбцах не одинаковое, что затруднит последующий анализ
#Так же можно предварительно оценить адекватность соотношения "параметр - тип данных"
```

```python
#Удаляем "лишние" (не влияющие на анализ) колонки

useless_columns = ['Name', 'Ticket', 'Cabin']
titanic_clean_1 = titanic_data.drop(columns=useless_columns)
```

```python
#Проверяем результат
titanic_clean_1.head()
```

```python
#imputing - заполнение пропущенных данных

def impute_data(df):
    df_copy = df.copy() #копируем, чтобы не изменять оригинал
    
    #заполняем пропущенные значения для категориальных переменных самым частым значением в колонке
    categorial_vars = list(df_copy.select_dtypes(include=['object']).columns.values)
    for col in categorial_vars:
        if (df_copy[col].isnull().values.any()):
            df_copy[col].fillna(df_copy[col].mode()[0], inplace=True)

    #заполняем пропущенные значения для количественных переменных медианой по столбцу
    numerical_vars = list(df_copy.select_dtypes(include=['int64', 'float64']).columns.values)
    for col in numerical_vars:        
        if (df_copy[col].isnull().values.any()):
            df_copy[col].fillna(df_copy[col].median(), inplace=True)
            
    return df_copy
```

```python
#Применяем функцию к "titanic_clean_1" и проверяем корректность выполнения
titanic_clean_2 = impute_data(titanic_clean_1)
titanic_clean_2.info()
```

```python
#В результате проделанных манипуляций мы получили новый датафрейм titanic_clean_2, содержащий "чистые" данные
```

```python
#Последним шагом этого раздела, для удобства дальнейшего анализа преобразуем данные о порте посадки (Embarked)
#и поле (Sex) из строкового типа в числовой и сохраним в новый датафрейм, оставив при этом старый вариант
#создадим словари
dic_1 = {'male':1, 'female':0}
dic_2 = {'S':1, 'C':2, 'Q':3}
titanic_clean_3 = titanic_clean_2.copy()
titanic_clean_3['Sex'] = titanic_clean_3['Sex'].apply(lambda x: dic_1[x])
titanic_clean_3['Embarked'] = titanic_clean_3['Embarked'].apply(lambda x: dic_2[x])
titanic_clean_3.head()
```

```python
#Воспользуемся t-SNE представлением данных
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

titanic_clean_3['Sex'] = pd.factorize(titanic_clean_3['Sex'])[0]
titanic_clean_3['Embarked'] = pd.factorize(titanic_clean_3['Embarked'])[0]
#смасштабируем выборку – из каждого признака вычтем его среднее и поделим на стандартное отклонение, это делает StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(titanic_clean_3)

tsne = TSNE(random_state=17)
tsne_representation = tsne.fit_transform(X_scaled)

plt.scatter(tsne_representation[:, 0], tsne_representation[:, 1]);
```
